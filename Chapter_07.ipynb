```python
# 標準ライブラリを利用したWebページの取得
# テキスト通りだとデータがgzip形式のためデコーディングエラー
import http.client

conn = http.client.HTTPSConnection('www.python.org')
conn.request('GET', '/downloads/')
response = conn.getresponse()
text = response.read().decode('UTF-8')
print(text)
conn.close()
```

```python
# 標準ライブラリを利用したWebページの取得(改修)
import http.client
import gzip                                             # 解凍用にgzipモジュールをインポート

conn = http.client.HTTPSConnection('www.python.org')
conn.request('GET', '/downloads/')
response = conn.getresponse()
decompressed_data = gzip.decompress(response.read())    # gzip解凍
text = decompressed_data.decode('UTF-8')
print(text)
conn.close()
```

```python
# 練習7-6
from random import randint
digit = 3   # 数字の桁数

print(f'数当てゲームをします。{digit}桁の数を当てて下さい！')

# 正解を作成
def answer_list():
  answer = list()
  for n in range(digit):
    answer.append(randint(0, 9))
  # print(f'{answer} (動作確認用に回答を表示)')
  return answer

# 予想の入力
def prediction_list():
  prediction = list()
  for n in range(digit):
    prediction.append(int(input(f'{n + 1}桁目の予想を入力(0～9) >> ')))
  return prediction

# ヒットアンドブロー
def hit_and_blow(answer_list, prediction_list):
  hit = 0
  blow = 0
  blow_answer_list = list()               # blow判定用リスト
  blow_prediction_list = list()           # blow判定用リスト
  for n in range(len(prediction_list)):   # ヒット数確認
    if prediction_list[n] == answer_list[n]:
      hit += 1
    else:                                 # blow判定用リストに追加
      blow_answer_list.append(answer_list[n])
      blow_prediction_list.append(prediction_list[n])
  for n in blow_prediction_list:          # blow判定用リストを用いてブロー数確認
    if n in blow_answer_list:
      blow += 1
      blow_answer_list.remove(n)
  return hit, blow

answer = answer_list()
is_play = True
while is_play == True:
  prediction = prediction_list()
  hit, blow = hit_and_blow(answer, prediction)
  if hit == digit:
    print('正解！', end = '')
    is_play = False
  else:
    print(f'{hit}ヒット！{blow}ブロー！')
    if int(input('続けますか？ (1:続ける / 2:終了) >> ')) != 1:
      print('残念！', end = '')
      is_play = False
print(f'数は「{"".join(map(str, answer))}」でした！')
print('ゲームを終了します。')
```
