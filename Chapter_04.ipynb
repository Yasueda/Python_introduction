```python
# 寝るまでひつじを数えるのを繰り返す
is_awake = True
count = 0
while is_awake == True:   # 起きている状態ならば
  count += 1
  print(f'ひつじが{count}匹')
  key = input('もう眠りそうですか？ (y/n) >> ')
  if key == 'y':
    is_awake = False      # 眠ったのでフラグをFalseにする
print('おやすみなさい')
```

```python
# 繰り返しを使って得点リストを作成する
count = 0                                         # カウンタ変数
student_num = int(input('学生の数を入力 >> '))    # 学生の数
score_list = list()                               # 得点リスト(空のリストを準備)
while count < student_num:                        # 入力された学生の数より小さければ繰り返す
  count += 1
  score = int(input(f'{count}人目の試験の得点を入力 >> '))
  score_list.append(score)                        # 入力され得点を得点リストに追加
print(score_list)
total = sum(score_list)
print(f'平均点は{total / student_num}点です')
```

```python
# リストの全要素を繰り返し参照する
scores = [80, 20, 75, 60]
count = 0
while count < len(scores):    # リストの要素数より小さければ繰り返す
  if scores[count] >= 60:
    print('合格')
  else:
    print('不合格')
  count += 1
```

```python
# forでリストの全要素を参照する
scores = [80, 20, 75, 60]
for data in scores:
  if data >= 60:
    print('合格')
  else:
    print('不合格')
```

```python
# データのまとまりからサンプルを抽出する
ages = [28, 50, 8, 20, 78, 25, 22, 10, 27, 33]    # 抽出データ
num = 5                                           # 目標の抽出数
samples = list()                                  # サンプルデータを格納するリスト
for age in ages:
  if 20 <= age < 30:
    if len(samples) < num:
      samples.append(age)
print(samples)
```

```python
# 目標数に達したら繰り返しを修了する
ages = [28, 50, 8, 20, 78, 25, 22, 10, 27, 33]    # 抽出データ
num = 5                                           # 目標の抽出数
samples = list()                                  # サンプルデータを格納するリスト
for age in ages:
  if 20 <= age < 30:
    if len(samples) < num:
      samples.append(age)
      if len(samples) == num:                     # 抽出数が目標に達したのでfor文を強制終了
        break
print(samples)
```

```python
# 不要な買いのループをスキップする
ages = [28, 50, 'ひみつ', 20, 78, 25, 22, 10, '無回答', 33]
samples = list()                  # サンプルデータを格納するリスト
for data in ages:
  if not isinstance(data, int):   # 整数でないデータはスキップ
    continue
  if data < 20 or data >= 30:     # 目的の条件に合致しないデータはスキップ
    continue
  samples.append(data)
print(samples)
```

```python
# 練習4-2

# 例外処理なし
count = 1
ans = 'y'
print('カレーを召し上がれ')
while ans == 'y':
  print(f'{count}皿のカレーを食べました')
  count += 1
  ans = input('おかわりはいかがですか？ (y/n) >> ')
print('ごちそうさまでした')

# 例外処理あり
count = 1
print('カレーを召し上がれ')
while True:
  print(f'{count}皿のカレーを食べました')
  key = input('おかわりはいかがですか？ (y/n) >> ')
  if key == 'y':
    count += 1
  elif key == 'n':
    break
  else:
    pass
print('ごちそうさまでした')

# 回答例
count = 1
ans = True
print('カレーを召し上がれ')
while ans:
  print(f'{count}皿のカレーを食べました')
  key = input('おかわりはいかがですか？ (y/n) >> ')
  if key == 'y':
    count += 1
  else:
    ans = False
print('ごちそうさまでした')
```

```python
# 練習4-5
temp = list()
start_time = 8          # 登録区間の時刻(開始)
end_time = 17           # 登録区間の時刻(終了)
inaccurate_time = 13    # 不正確データの時刻

for num in range(start_time, end_time + 1):
  temp.append(float(input(f'{num}時の温度を入力 >> ')))
print('temp')
for num in range(len(temp)):
  print(f'{num + start_time}時：{temp[num]}度； ', end = '')
print('\n', end = '')

temp_new = list()
for num in range(len(temp)):
  if num + start_time == inaccurate_time:
    temp_new.append('N/A')
  else:
    temp_new.append(float(temp[num]))

total = 0
count = 0
print('temp_new')
for num in range(len(temp_new)):
  print(f'{num + start_time}時：{temp_new[num]}度； ', end = '')
  if isinstance(temp_new[num], float):
    total += temp_new[num]
    count += 1
print('\n', end = '')
print(f'平均気温は{(total / count):.2f}度です')
```

```python
# 練習4-6
numbers = [1, 1]
num = sum(numbers)
count = 1
while num <= 1000:
  numbers.append(num)
  num += numbers[count]
  count += 1
print(f'numbers = {numbers}')

ratios = list()
digit = 3                                       # 小数点以下桁数
for count in range(len(numbers) - 1):
  num = numbers[count + 1] / numbers[count]
  num = int(num * 10 ** digit) / 10 ** digit    # 桁数調整
  ratios.append(num)
print(f'ratios = {ratios}')
```
