```python
# 勇者を表すクラスの定義と利用
class Hero:   # データとしてnameとhp、関数としてsleepを持つオリジナル設計図Heroを定義
  name = '松田'
  hp = 100
  def sleep(self, hours):
    print(f'{self.name}は{hours}時間寝た！')
    self.hp += hours

# ゲーム開始
print('スッキリファンタジーXII ～金色の理想郷～')
h = Hero()    # HP100の勇者・松田がオブジェクトとして誕生
h.sleep(3)
print(f'{h.name}のHPは現在{h.hp}です')
```

```python
# オブジェクトのidentityを確認
scores1 = [80, 40, 50]
scores2 = [80, 40, 50]
print(f'scores1のidentity: {id(scores1)}')
print(f'scores2のidentity: {id(scores2)}')

if scores1 == scores2:            # 等価判定
  print('scores1とscores2は同じ内容です')
else:
  print('scores1とscores2は違う内容です')

if id(scores1) == id(scores2):    # 等値判定
  print('scores1とscores2は同じ存在です')
else:
  print('scores1とscores2は違う存在です')
```

```python
# リストオブジェクトのコピーによる不可解な動作
scores1 = [80, 40, 50]
scores2 = [80, 40, 50]
print(f'scores1の先頭要素は{scores1[0]}')
print(f'scores2の先頭要素は{scores2[0]}')
print(f'scores1のidentity: {id(scores1)}')
print(f'scores2のidentity: {id(scores2)}')

print('変数scores2の中身を変数scores1に代入(コピー)します')
scores1 = scores2
print(f'scores1のidentity: {id(scores1)}')
print(f'scores2のidentity: {id(scores2)}')

print('scores1の先頭要素を90に書き換えます')
scores1[0] = 90

print(f'90を代入したscores1の先頭要素は{scores1[0]}')
print(f'90を代入していないscores2の先頭要素は{scores2[0]}')
print(f'scores1のidentity: {id(scores1)}')
print(f'scores2のidentity: {id(scores2)}')
```

```python
# 関数に変数の内容を書き換えられてしまう
def add_suffix(names):                    # 渡されたリスト内の名前に「さん」を付ける関数
  for i in range(len(names)):
    names[i] = names[i] + 'さん'
  return names

before_names = ['松田', '浅木', '工藤']
after_names = add_suffix(before_names)
print('さん付け後: ' + after_names[0])    # add_suffixから返されたリスト内の先頭の要素
print('さん付け前: ' + before_names[0])   # add_suffixに渡したリスト内の先頭の要素
```

```python
# 防御的コピーを用いて悪影響を防ぐ
def add_suffix(names):    # 渡されたリスト内の名前に「さん」を付ける関数
  for i in range(len(names)):
    names[i] = names[i] + 'さん'
  return names

before_names = ['松田', '浅木', '工藤']
copied_names = list()
for n in before_names:    # リストを複製する
  copied_names.append(n)
after_names = add_suffix(copied_names)
print('さん付け後: ' + after_names[0])
print('さん付け前: ' + before_names[0])
print('複製リスト: ' + copied_names[0])
```

```python
# 文字列を渡しても悪影響が起きない
def add_suffix(name):                   # 渡された名前に「さん」を付ける関数
  name = name + 'さん'
  return name

before_name = '松田'
after_name = add_suffix(before_name)    # 文字列を渡している
print('さん付け後: ' + after_name)      # add_suffixから返された名前
print('さん付け前: ' + before_name)     # add_suffixに渡した名前
```

```python
# identity値の変化の比較
"""
オブジェクトが不変となる代表的な型
・int型, str型, bool型などの一部の標準的な型
・tuple型(コンテナのタプル)
"""
print('identityの変化を比較')

names = list()    # リストの場合
print(f'list(変更前): {id(names)}')
names.append('松田')
print(f'list(変更後): {id(names)}')

name = '松田'     # 文字列の場合
print(f'str(変更前): {id(name)}')
name = 'スーパー' + name
print(f'str(変更後): {id(name)}')
```

```python
# 練習6-3
def welcome(u):
  print(f'ようこそ{u["name"]}さん')
  u['age'] = u['age'] + 1
  print(f'あなたは来年{u["age"]}歳だから大吉です！')

username = input('名前を入力してください >> ')
userage = int(input('年齢を入力してください >> '))
user = {'name': username, 'age': userage}
copied_user = user.copy()
welcome(copied_user)
print(f'{user["age"]}歳の{user["name"]}さん、またプレイしてくださいね')
```
