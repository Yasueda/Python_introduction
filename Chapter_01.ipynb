```python
# print関数で数値を書く
print(1)
print(10)
```

```python
# 加算と減算
print(1 + 1)
print(10 - 2)

# 算術演算子の種類
"""
演算子の優先順位
高  **
中  * / %
低  + -
"""
# 足し算（加算）
print('print(10 + 10) =', 10 + 10)
# 引き算（減算）
print('print(10 - 1) =', 10 - 1)
# 掛け算（乗算）
print('print(2 * 3) =', 2 * 3)
# 割り算（除算）
print('print(7 / 4) =', 7 / 4)
# 割り算の商
print('print(7 // 4) =', 7 // 4)
# 割り算の余り
print('print7 % 4) =', 7 % 4)
# べき乗（累乗）
print('print(2 ** 3) =', 2 ** 3)
```

```python
# 文字列の演算
print('1' + '1')

# 文字列における算術演算子
# 文字列の連結
print('print(\'1\' + \'1\') =', '1' + '1')
# 文字列の反復
print('print(\'オラ\' * 3) =', 'オラ' * 3)
```

```python
# 文字列の演算の応用
print('Python' + 'の世界へようこそ')
print('Pythonは' + 'とっても' * 3 + '楽しいですよ')
```

```python
"""
代表的なエスケープシーケンス
\n  改行を表す制御文字
\\  \(バックスラッシュ)
\'  '(シングルクォーテーション)
\"  "(ダブルクォーテーション)
"""

# 文字列の途中で改行する
print('はじめまして松田です身体を動かすのが好きです')
print('はじめまして\n松田です\n身体を動かすのが好きです')
print('引用符には、\'と\"があります')
```

```python
# ヒアドキュメントの利用
print('''はじめまして
松田です
身体を動かすのがすきです''')
```

```python
# さらに改行を追加すると…
print('''
はじめまして
松田です
身体を動かすのがすきです
''')
```

```python
# 前後の改行を取り除く
print('''
はじめまして
松田です
身体を動かすのが好きです
'''.strip())
```

```python
# 変数を利用してコードを改善
print('半径が3cmの円の直径は、')
dia = 3 * 2         # diaはdiameter(直径)の略
print(dia)          # 計算結果を利用(参照)
print('その円の円周の長さは、直径×円周率で求まるため、')
print(dia * 3.14)   # 計算結果を再利用(参照)
```

```python
"""
変数名のルール
1,  予約語は使用できない
    > import keyword
    > print(keyword.kwlist)
2,  先頭の文字に数字は指定できない
3,  先頭にアンダースコアを2つ付けた名前は原則として使用しない
4,  大文字／小文字、全角／半角は区別される
5,  小文字で始まるわかりやすい名前が望ましい
    複数の単語から作る識別子の命名規則はPythonではスネークケースを推奨している
ex, 中身を絶対に書き換えられたくない変数は、目立つように大文字の変数名を付ける習慣がある(「TAX_RATE」など)
"""

# 変数の上書き
count = 3   # 変数countに3を代入
print('今日カレーを食べた回数は')
print(count)
count = 5   # 変数countに5を代入
print('うそ。本当は')
print(count)
```

```python
# 変数の現在の値に加減乗除する
"""
主な複合代入演算子
+=  右辺の値と左辺の変数の値を足し算して変数に代入
-=  右辺の値と左辺の変数の値を引き算して変数に代入
*=  右辺の値と左辺の変数の値を掛け算して変数に代入
/=  右辺の値と左辺の変数の値を割り算して変数に代入
"""

age = 24
age += 1
print(age)

price = 2600
price *= 1.5
print(price)
```

```python
"""
主なデータ型
int   整数    100, -100
float 少数    3.14, -0.5
str   文字列  "Hello", "カレー"
bool  真偽値  True, False
"""

# 変数には異なるデータ型の値を代入できる
x = '松田'  # str型の値を代入
print(x)
x = 23      # int型の値を代入
print(x)
x = 175.6   # float型の値を代入
print(x)
```

```python
"""
データ型変換のための命令
int関数   int(x)    変数xの値を整数に変換(xが小数の場合は小数点以下を切り捨て。数値として解釈できない文字列の場合はエラー)
float関数 float(x)  変数xの値を小数に変換(数値として解釈できない文字列の場合はエラー)
str関数   str(x)    変数xの値を文字列に変換
bool関数  bool(x)   変数xの値を真偽値に変換
"""

# データ型の変換
x = 3.14
y = int(x)      # 変数xの値をint型(整数)に変換した結果を変数yに代入
print(y)        # 変換結果を表示
print(type(y))  # 返還後のデータ型を表示
z = str(x)      # 変数xの値をstr型(文字列)に変換した結果をzに代入
print(z)        # 変換結果を表示
print(type(z))  # 返還後のデータ型を表示
print(z * 2)
```

```python
# 割り勘計算プログラムの修正
price = input('料金を入力 >>')                  # キーボード入力結果はstr型
price = int(price)                              # strからintへ変換
number = input('人数を入力 >>')                 # キーボード入力結果はstr型
number = int(number)                            # strからintへ変換
payment = price / number                        # 割り算の結果はfloat型
payment = int(payment)                          # floatからintへ変換(小数点以下切り捨て)
print('お支払いは' + str(payment) + '円です')   # 演算は、文字列同士か数値同士で
```
