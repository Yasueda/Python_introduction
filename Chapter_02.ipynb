```python
# 点数管理プログラム
network = 88                            # ネットワークは88点
database = 95                           # データベースは95点
security = 90                           # セキュリティは90点
total = network + database + security   # 合計を計算
avg = total / 3                         # 平均を計算
print(f'合計点：{total}')
print(f'平均点：{avg}')
```

```python
# リストの合計値と平均値を求める
scores = [88, 90, 95]
total = sum(scores)
avg = total / len(scores)   # len関数でリストの要素数を求める
print(f'合計：{total}点、平均点：{avg}点')
```

```python
# ディクショナリの作成
"""
キーには、文字列のほか数値型など、さまざまな型のデータを指定できる
キーのデータ型は要素ごとにことなってもよい
キーの重複も許される(ただし、最後に指定したもの以外は無視される)
"""
scores = {'network':60, 'database':80, 'security':50}   # ディクショナリの作成
print(scores)                                           # ディクショナリの全要素を表示
```

```python
# ディクショナリの要素を参照
scores = {'network':60, 'database':80, 'security':50}
print(scores['database'])   # ディクショナリの要素を指定
```

```python
# セットの利用
"""
1, 重複した値を格納できない
2, 添え字やキーの概念がなく、特定の要素に対して代入・参照する方法が存在しない
3, 添え字がないため、要素は順序を持たない
4, append関数ではなくadd関数で要素を追加する
ex,セットは中括弧、配列は大括弧！
"""

scores = {70, 80, 55, 80}
scores.add(80)    # 重複する値を格納できないため、80は1つしか格納されていない
print(scores)     # 順序を持たないため、記述順とは異なる順で表示される
print(f'要素数は{len(scores)}')
print(f'合計は{sum(scores)}')
```

```python
# コンテナの相互変換
scores = {'network':60, 'database':80, 'security':60}
members = ['松田', '浅木', '工藤']
print(tuple(members))         # リストmembersをタプルに変換して表示(順序関係がないセットから変換すると順序は保証されない)
print(list(scores))           # scoresのキーをリストに変換して表示(順序関係がないセットから変換すると順序は保証されない)
print(set(scores.values()))   # scoresの値をセットに変換して表示(重複が許されるリストやタプルからは、重複が排除される)
```

```python
# ディクショナリへの変換
keys = ['network', 'database', 'security']
scores = [60, 80, 60]
print(dict(zip(keys, scores)))    # キーを格納したリストと値を格納したリストが必要
```

```python
# 2次元リストの例
a = [1, 2, 3]
b = [4, 5, 6]
c = [a, b]        # aを0番目、bを1番目とする2次元リストcを定義

print(c)          # リストc全体を参照
print(c[0])       # リストcの0番目(リストa)だけを表示
print(c[1][2])    # リストcの1番目(リストb)の2番目だけを表示
```

```python
# 2人の共通点を探す
member_hobbies = {
    '松田': {'SNS', '麻雀', '自転車'},
    '浅木': {'麻雀', '食べ歩き', '数学', '数学', '数学'}
}
common_hobbies = member_hobbies['松田'] & member_hobbies['浅木']
# 2人に共通する趣味一覧を表示する
print(common_hobbies)
```

```python
# 4つの集合演算
A = {1, 2, 3, 4}
B = {2, 3, 4, 5}
print(A | B)    # 和集合
print(A & B)    # 積集合
print(A - B)    # 差集合
print(A ^ B)    # 対称差
```

```python
# 練習2-2
print("""国語・算数・理科・社会・英語の合計点と平均点を計算します。
各教科の点数を入力してください。""")
scores = []
scores.append(int(input('国語の点数 >> ')))
scores.append(int(input('算数の点数 >> ')))
scores.append(int(input('理科の点数 >> ')))
scores.append(int(input('社会の点数 >> ')))
scores.append(int(input('英語の点数 >> ')))
print(f'合計点：{sum(scores)}点、平均点：{sum(scores) / len(scores)}点')
```

```python
# 練習2-3
a_hobbies = {'読書', '昼寝', '映画鑑賞', '散歩', '料理'}
b_hobbies = {'テニス', '将棋', '料理', '読書', '旅行'}
input('心の準備ができたらEnterキーを押してください')
intersection = a_hobbies & b_hobbies
union = a_hobbies | b_hobbies
compatibility_rate = len(intersection) / len(union) * 100
print(f'2人の趣味の相性度は... {compatibility_rate}% です！')
```
