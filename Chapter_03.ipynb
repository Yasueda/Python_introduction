```python
# 答えが分岐するチャットボット
"""
if文の基本構文
if 条件式:
  条件が成立したときの処理(ifブロック)
else:
  条件が成立しなかったときの処理(elseブロック)
"""
name = input('あなたの名前を教えてください >> ')
print(f'{name}さん、こんにちは')
food = input(f'{name}さんの好きな食べ物を教えてください >> ')
if food == 'カレー':  # 変数foodがカレーだったら
  print('素敵です。カレーは最高ですよね！！')
else:
  print(f'私も{food}が好きですよ')
```

```python
# 常に追試をうけることになる判定プログラム
score = int(input('試験の点数を入力してください >> '))
if score >= 60:   # 変数scoreが60以上だったら
  print('合格！')
  print('よくがんばりましたね')
else:
  print('残念ながら不合格です')
print('追試を受けてください')
```

```python
# どんなカレーでも絶賛するチャットボット
name = input('あなたの名前を教えてください >> ')
print(f'{name}さん、こんにちは')
food = input(f'{name}さんの好きな食べ物を教えてください >> ')
if 'カレー' in food:    # 変数foodにカレーが含まれているか
  print('素敵です。カレーは最高ですよね！')
else:
  print(f'私も{food}が好きですよ')
```

```python
# 100点があるかどうかを調べる
scores = [80, 100, 20, 60]    # 試験結果のリスト
if 100 in scores:
  print('100点満点の試験があったんですね。おめでとう！')
else:
  print('次はどれか1つでも100点満点をとろう')
```

```python
# ディクショナリのキーをチェックする
scores = {'network': 60, 'database': 80, 'security': 50}    # 試験結果のディクショナリ
key = input('追加する科目名を入力してください >> ')
if key in scores:
  print('すでに登録済です')
else:
  data = int(input('得点を入力してください >> '))
  scores[key] = data
print(scores)
```

```python
# elseブロックのない分岐
name = input('あなたの名前を教えてください >> ')
print(f'{name}さん、こんにちは')
if name == '松田':
  print('松田さんに会えてうれしいです')
# nameが松田ではないときの処理はない
food = input(f'{name}さんの好きな食べ物を教えてください >> ')
if 'カレー' in food:
  print('素敵です。とにかくカレーは最高ですよね!!')
else:
  print(f'私も{food}が好きですよ')
```

```python
# 多分岐するif文
score = int(input('試験の点数を入力してください >> '))
if score < 0 or score > 100:    # 条件式1
  print('異常な特典です')
  print('入力し直してください')
elif score >= 60:               # 条件式2
  print('合格')
  print('よくがんばりましたね')
else:                           # これまでの条件式がすべてFalseだったら実行される
  print('残念ながら不合格です')
  print('追試を受けてください')
```

```python
# 晩ご飯をレコメンドするチャットボット
print('すべての質問に y または n で答えてください')
okane_aruka = input('お金に余裕はありますか？ >> ')
if okane_aruka == 'y':            # if-else構文
  onaka_suiteruka = input('お腹がすごく空いていますか？ >> ')
  nomitai_kibunka = input('ビールを飲みたいですか？ >> ')
  if onaka_suiteruka == 'y' and nomitai_kibunka == 'y':
    print('焼き肉はいかがですか')
  elif onaka_suiteruka == 'y':    # if-else構文
    print('カレーはいかがですか')
  elif nomitai_kibunka == 'y':    # if-else構文
    print('焼き鳥はいかがですか')
  else:                           # if-else構文
    print('パスタはいかがですか')
  yashoku_iruka = input('夜食は必要ですか？ >> ')
  if yashoku_iruka == 'y':        # ifのみの構文
    print('コンビニのチキンはいかがですか')
else:                             # if-else構文
  print('家で食べましょう')
```

```python
# 練習3-3-(1)
isError = input('isError? (True/False) >> ')
n = int(input('n? (int) >> '))
if isError == 'False' and n < 100:
  print('\'isError == False\' and \'n < 100\'')
```

```python
# 練習3-3-(2)
num = int(input('偶奇判定：数値を入力してください >> '))
if num % 2 == 0:
  print(f'{num}は偶数です')
else:
  print(f'{num}は奇数です')
```

```python
# 練習3-3-(2) 三項条件演算子
num = int(input('偶奇判定：数値を入力してください >> '))
div = '偶数' if num % 2 == 0 else '奇数'
print(f'{num}は{div}です')
```
